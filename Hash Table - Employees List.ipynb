```python
#We define a default basic function for generating a hash identifier --> Adds all digits together
def phoneFunction(number):
    identifier = 0
    for digit in number:
        identifier += int(digit)
    return identifier


#This is a general CONTACT class:
# - Name
# - Number
# - Address
class contact:
    def __init__(self, name, number, address = None):
        self.name = name
        self.number = number
        self.address = address

    def getInfo(self):
        return [self.name, self.number, self.address]
        
    def getIdentifier(self):
        return self.number

    
#This is a general HASHTABLE CLASS
# - __init__: Generates an string of electable length (specific hash function can be included as well)
# - hashFunc: Generates the hash number (adjusted to the string length) based on the identifier given by the function selected above
# - insertElem: Allows to insert the element desired (identifier string must be given as well)
# - getElem: Given the string identifier returns all the contact data for the correspondent element
class hashTable:
    def __init__(self, num_options, function = phoneFunction):
        self.opt = num_options
        self.array = [None]*self.opt
        self.func = function
    
    #INPUT: Numerical Identifier ('Unique' number)
    #OUTPUT: Hash number (for the sel.array)
    def hashFunc(self, value):
        val = self.func(value)
        hash_num = val%self.opt
        
        return hash_num
    
    
    #INPUTS: 1) Element (a class)
    #        2) Unique numerical identifier
    #OUTPUT: Bool --> TRUE __ if insertion correct | FALSE __ if there is a problem during insertion
    def insertElem(self, element, id_val):
        hash_num = self.hashFunc(id_val)
        
        if self.array[hash_num] == None:
            self.array[hash_num] = element
            return True
        
        return False
    
    #INPUT: Unique numerical identifier (from element desired)
    #OUTPUT: Element desired
    def getElem(self, id_val):
        hash_num = self.hashFunc(id_val)
        
        if self.array[hash_num] == None:
            return False
        else: #type(self.array[identifier]) == contact:
            return self.array[hash_num]
        #else:
            #WE HAVE TO MANAGE POSSIBLE COLLISIONS
            #*************************************
        #    return False


#This function gets a string of element with class "contact" and inserts them in the Hash Table provided
#INPUTS: 1) Hash table --> "hashTable"
#        2) String of elements with class "contact"
#OUTPUS: Bool --> TRUE __ Insertion is correct | FALSE __ Insertion is not correct
def insertNames(table, users):
    inserted = True
    for user in users:
        inserted = table.insertElem(user,user.getIdentifier())
        if not inserted:
            return False
    return True


if __name__ == "__main__":
    
    table = hashTable(250)
    
    users = [["Juan","7797778811","21 Fell Street, SF, CA"]
             ,["Maria","7793331122","21 Market Street, SF, CA"]
             ,["Adam Smith","8993331133","24 Castro Street, SF, CA"]]
    
    users_list = []
    for user in users:
        user_class = contact(user[0],user[1],user[2])
        users_list.append(user_class)
    
    insertNames(table,users_list)
    
    for user in users_list:
        print(table.getElem(user.getIdentifier()).getInfo())
```
