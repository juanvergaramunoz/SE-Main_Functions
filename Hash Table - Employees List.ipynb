```python
from LinkedList_Class import linkedList
from HashTable_Class import hashTable

##---------------------------##
## Aux General Contact Class ##
##---------------------------##
#
# INCLUDES THE FOLLOWING INFO:
# - X.name
# - X.number
# - X.address
#
##----- FUNCTIONS FOR THIS CLASS -----##
#
# - __init__: Generates an element with the provided info
#
# - getInfo: Returns a list with the details for that contact
#
# - getIdentifier: Returns the unique identifier (we have chosen the phone number)
#
class contact:
    def __init__(self, name, number, address = None):
        self.name = name
        self.number = number
        self.address = address

    def getInfo(self):
        return [self.name, self.number, self.address]
        
    def getIdentifier(self):
        return self.number



#This function gets a string of element with class "contact" and inserts them in the Hash Table provided
#INPUTS: 1) Hash table --> "hashTable"
#        2) String of elements with class "contact"
#OUTPUS: Bool --> TRUE __ Insertion is correct | FALSE __ Insertion is not correct
def insertNames(table, users):
    inserted = True
    for user in users:
        inserted = table.insertElem(user,user.getIdentifier())
        if not inserted:
            return False
    return True


if __name__ == "__main__":
    
    #We iniate the hashTable to have 100 positions (10 digit phones can add up to 90 different values)
    table = hashTable(100)
    
    #We generate our set of users (a list of lists)
    users = [["Juan","7777778811","21 Fell Street, SF, CA"]
             ,["Maria","7973112233","22 Market Street, SF, CA"]
             ,["Bea","7379112233","23 Polk Street, SF, CA"]
             ,["Adam Smith","9991122233","24 Castro Street, SF, CA"]]
    
    #We create a CONTACT class from each user list of values
    users_list = []
    for user in users:
        user_class = contact(user[0],user[1],user[2])
        users_list.append(user_class)
    
    #We call the function defined above to iterate over each element and insert them into the HASHTABLE
    insertNames(table,users_list)
    
    #WE PRINT EACH OF THE VALUES ADDED
    print("**** USERS INSERTED ****")
    for user in users_list:
        user_ID = user.getIdentifier()
        user_stored = table.getElem(user_ID)
        print(user_stored.getInfo())
    
    #WE ERASE USER AT POSITION <POS> FROM THE LIST OF ELEMENTS WITH CLASS 'CONTACT' (users_list)
    print("\n**** USER TO BE ERASED ****")
    pos = 1
    user = users_list[pos]
    user_ID = user.getIdentifier()
    print("We are about to delete the user '",user.getInfo()[0],"' with phone = ",user_ID," (stored on position",pos+1,"from vector 'users_list')")
    success = table.deleteElem(user_ID)
    if success:
        print("- DELETION OF",user.getInfo()[0],"HAS BEEN COMPLETED SUCCESSFULLY")
        
        #WE NOW CHECK IF THERE IS STILL INFORMATION ABOUT THE ERASED USER
        print("\nNOW, we check if user '",user.getInfo()[0],"' with phone = ",user_ID,", STILL EXISTS inside the hashTable")
        success = table.getElem(user_ID)
        if success:
            print("  <ERROR!> ELEMENT WITH NAME '",user.getInfo()[0],"' HAS BEEN FOUND SUCCESSFULLY")
        else:
            print("- ELEMENT '",user.getInfo()[0],"' HAS NOT BEEN FOUND. Deletion successful!!")
    else:
        print("  <ERROR!> DELETION OF '",user.getInfo()[0],"' HAS BEEN COMPLETED UNSUCCESSFULLY. Element not found in hashTable")
    
    
```
